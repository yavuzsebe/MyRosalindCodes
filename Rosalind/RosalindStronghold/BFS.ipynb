```python
with open("/Users/yavuzsebe/Downloads/rosalind_bfs_data.txt","r") as inputSet:   #sample output: 2 4 2 2 2 2
    inputSet=inputSet.readlines()
    i=0
    while i<len(inputSet)-1:
        inputSet[i]=inputSet[i][0:-1]
        i+=1
    i=1

root=1
vertices=int(inputSet[0].split()[0])

edgeList=[]
while i<len(inputSet):
    edgeList.append(inputSet[i])
    i+=1

splitList=[]
for i in edgeList:
    splitList.append(i.split())

splitList.sort()
graphKeys=[]
for i in splitList:
    if int(i[0]) not in graphKeys:
        graphKeys.append(int(i[0]))
        
graphDict={}
for i in splitList:
    if int(i[0]) not in graphDict.keys():
        graphDict[int(i[0])]=[]
    if int(i[1]) not in graphDict[int(i[0])]:
        graphDict[int(i[0])]+=[int(i[1])]

Q=[] #queue
Q.append(root)
distance=1
V=[] #visited nodes
Qq=[]
dictDistance={1: [0]}
while len(V)<vertices:
    if Q[0] in graphDict and graphDict[Q[0]] not in Q:
        Qq=graphDict[Q[0]]
        
        for i in Qq:
            if i not in dictDistance:
                dictDistance[i]=[distance]
        distance+=1
    V.append(Q[0])
    Q.pop(0)
    Q+=Qq

answer=[]
for i in range(vertices):
    answer.append(-1)
for i in dictDistance:
    answer[i-1]=dictDistance[i][0]

answerStr=""

for i in answer:
    answerStr+=str(i)+" "

print(answerStr)
```

```python
with open("/Users/yavuzsebe/Downloads/rosalind_bfs.txt","r") as inputSet:
    inputSet = inputSet.readlines()
    i = 0
    while i < len(inputSet) - 1:
        inputSet[i] = inputSet[i].strip()
        i += 1
    i = 1

root = 1
vertices = int(inputSet[0].split()[0])

edgeList = []
while i < len(inputSet):
    edgeList.append(inputSet[i])
    i += 1

splitList = []
for i in edgeList:
    splitList.append(i.split())

graphDict={}
for i in splitList:
    if int(i[0]) not in graphDict.keys():
        graphDict[int(i[0])]=[]
    if int(i[1]) not in graphDict[int(i[0])]:
        graphDict[int(i[0])]+=[int(i[1])]

Q=[root] 
V=set() 
distances=[-1]*vertices
distances[root-1]=0

while Q:
    current=Q.pop(0)
    V.add(current)

    if current in graphDict:
        for neighbor in graphDict[current]:
            if neighbor not in V and distances[neighbor-1]==-1:
                distances[neighbor-1] = distances[current-1]+1
                Q.append(neighbor)

answerStr=" ".join(map(str, distances))
print(answerStr)
```
